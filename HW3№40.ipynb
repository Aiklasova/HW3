Сначала импортируем все нужные для работы библиотеки

```python
import numpy as np
import random
import matplotlib.pyplot as plt
```

Затем добавляем в наш массив все имена для подписи

```python
names = ['NASNet']
for i in range(8):
    names.append('Arch' + str(i+1))
```

Задаем диапазон для оси абсцисс

```python
x = np.linspace(100, 300, 5)
```

Заполняем массив Y рандомными значениями и добавляем графики в нашу фигуру
Затем заполняем подписи нашим массивом names и берем их в красивую рамку

```python
fig, ax = plt.subplots()
for i in range(8):
    y = [random.uniform(0.8, 1) for i in x]
    ax.plot(x, y, linestyle='solid')
plt.xlabel('epochs')
plt.ylabel('accurancy')
plt.title('Pseudorandom diagram')
lgnd = ax.legend(names, loc='upper left', shadow=True)
lgnd.get_frame().set_facecolor('#ccccef')
```
